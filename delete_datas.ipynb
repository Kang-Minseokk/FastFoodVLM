```python
from google.colab import drive
import os

# Google Drive 마운트
drive.mount('/content/drive')

def delete_files_with_keyword(paths, keyword='hamburger'):
    """여러 경로에서 키워드가 포함된 파일들을 삭제"""
    total_deleted = 0
    all_deleted_files = []

    for path in paths:
        print(f"\n경로: {path}")
        deleted_count = 0

        if not os.path.exists(path):
            print("디렉토리 없음")
            continue

        # 모든 하위 디렉토리 포함해서 검색
        for root, dirs, files in os.walk(path):
            for file in files:
                if keyword.lower() in file.lower():
                    file_path = os.path.join(root, file)
                    try:
                        os.remove(file_path)
                        all_deleted_files.append(file)
                        deleted_count += 1
                        print(f"✓ {file}")
                    except Exception as e:
                        print(f"✗ {file}: {e}")

        print(f"이 경로에서 {deleted_count}개 삭제")
        total_deleted += deleted_count

    print(f"\n총 {total_deleted}개 파일 삭제됨")
    return {"total": total_deleted, "files": all_deleted_files}

def preview_files(paths, keyword='hamburger'):
    """삭제 전 미리보기"""
    total_found = 0

    for path in paths:
        print(f"\n경로: {path}")
        found_count = 0

        if not os.path.exists(path):
            print("디렉토리 없음")
            continue

        for root, dirs, files in os.walk(path):
            for file in files:
                if keyword.lower() in file.lower():
                    print(f"  {file}")
                    found_count += 1

        print(f"→ {found_count}개 발견")
        total_found += found_count

    print(f"\n총 {total_found}개 파일 발견")
    return total_found

# 사용 예시
if __name__ == "__main__":
    # 경로 설정
    paths = [
        "/content/drive/Shared drives/달튼/YOLOv5/yolov5-master/data/food_image/images",
        "/content/drive/Shared drives/달튼/YOLOv5/yolov5-master/data/food_image/labels"
    ]

    # 미리보기
    found_count = preview_files(paths, 'hamburger')

    if found_count > 0:
        confirm = input(f"{found_count}개 파일을 삭제할까요? (y/N): ")
        if confirm.lower() == 'y':
            result = delete_files_with_keyword(paths, 'hamburger')
        else:
            print("취소됨")
    else:
        print("삭제할 파일 없음")
```
